# **Transformaciones**

```python
#Llenar los valores vacíos (NAN) de Description con 'No Description'
df['Description'] = df['Description'].fillna('No Description', inplace=True)
df.tail()
```

```python
#Reempazar los caracteres '£' por "" de las columnas Price (excl. tax), Price (incl. tax), Tax
df.loc[df['Price (excl. tax)'].str.contains('£', na = False), 'Price (excl. tax)'] = df['Price (excl. tax)'].str.replace('£', '')
df.loc[df['Price (incl. tax)'].str.contains('£', na = False), 'Price (incl. tax)'] = df['Price (incl. tax)'].str.replace('£', '')
df.loc[df['Tax'].str.contains('£', na = False), 'Tax'] = df['Tax'].str.replace('£', '')
```

```python
#Extracción de la cantidad numérica de stock de la cadena de texto
stock_part = df['Availability'].str.split('(', expand=True)[1]
number_string = stock_part.str.split(' ', expand=True)[0]

#Convertir a entero
df['Stock Quantity_No_RE'] = number_string.astype(int)
```

```python
# Elimina la columna original y renombrar la nueva
df.drop('Availability', axis=1, inplace=True)
# Limpieza/normalización de texto
df.rename(columns={'Stock Quantity_No_RE': 'Availability'}, inplace=True)
```

```python
df['Product Type'].unique()
```

```python
#Eliminaremos las columnas de Product Type, Tax y #Reviews ya que no aporta información valiosa
df.drop('Product Type', axis=1, inplace=True)
df.drop('Tax', axis=1, inplace=True)
df.drop('# Reviews', axis=1, inplace=True)
#Eliminamos uno de los Price ya que contienen la misma información
df.drop('Price (incl. tax)', axis=1, inplace=True)
```

# **Outliers**

```python
column_to_check = 'Price (excl. tax)'

# Calcular Q1, Q3 e IQR
Q1 = df[column_to_check].quantile(0.25)
Q3 = df[column_to_check].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Encontrar los outliers
outliers = df[(df[column_to_check] < lower_bound) | (df[column_to_check] > upper_bound)]

print(f"Total de Outliers detectados: {outliers.shape[0]}")
```
